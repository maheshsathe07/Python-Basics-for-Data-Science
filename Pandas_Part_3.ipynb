### merge and join operation

```python
data1 = pd.DataFrame({'key1':[1,2,3,4,5,6],
                      'key2':[4,5,6,7,8, 9],
                      'key3':[3,5,6,7, 9, 10]
}
)
```

```python
data2 = pd.DataFrame({'key1':[1,2,3,14,15,16],
                      'key4':[24,25,26,27,28, 29],
                      'key5':[13,15,36,27, 29, 10]
}
)
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1, data2) #it merges only that values in which key1 value is same in both data1 and data2 it is inner
```

```python
#only get common values
```

```python
pd.merge(data1, data2, how='left') #keeps data of left and common data of right here left is data1 
#in case of merge - always check resp. cols values.
```

```python
pd.merge(data1, data2, how='right')
```

```python
pd.merge(data1, data2, how='outer')
```

```python
pd.merge(data1, data2, how='cross')
```

```python
pd.merge(data1, data2, how = 'left', left_on = 'key2', right_on = 'key4') #based on match with key 2 and key4
```

```python
data1 = pd.DataFrame({'key1':[1,2,3,4,5,6],
                      'key2':[4,5,6,7,8, 9],
                      'key3':[3,5,6,7, 9, 10]},
                     index=['a', 'b', 'c', 'd', 'e', 'f']
)
```

```python
data2 = pd.DataFrame({'key11':[1,2,3,14,15,16],
                      'key22':[24,25,26,27,28, 29],
                      'key33':[13,15,36,27, 29, 10]
}, index=['a','b','h','i','j','k']
)
```

```python
data1
```

```python
data2
```

```python
data1.join(data2) #by default 'left' #in case of join is trying to check row index match
```

```python
data1.join(data2,how='inner')
```

```python
data1.join(data2,how='outer')
```

```python
data1.join(data2,how='right')
```

```python
data1.join(data2,how='cross')
```

### using .apply()

```python
#create a new col and put data where u convert fare value into a inr and keep that value in col
df["Fare_inr"]=df["Fare"].apply(lambda x : x*80) #function written in apply is applied on every value of fare
```

```python
df.head()
```

```python
df["len_name"] = df['Name'].apply(len)
```

```python
df
```

```python
#instead of using lambda function we can also do like this-
def inr_con(x):
    return x*80

df['Fare_inr']=df['Fare'].apply(inr_con)
```

```python
df.head()
```

```python
#separating in category as per Fare 
def create_flag(x):
    if x<10:
        return "cheap"
    
    elif x>=10 and x<20:
        return "mid"
    
    else:
        return "high"
```

```python
df["flag_fare"] = df['Fare'].apply(create_flag)
```
